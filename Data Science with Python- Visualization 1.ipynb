# Matplotlib

Matplotlib is a python 2D plotting library which produces graphs and figure across platforms. Matplotlib tries to make easy and hard things possible. 

For simple plotting the pyplot module provides a MATLAB-like interface, particularly when combined with Jupyter Notebooks.

Plotting requires action on a range of levels:
- color grading
- axis orientation
- labeling the data

The purpose of a plotting package is to assist you in visualizing your data as easily as possible, with all the necessary control.

Therefore, everything in matplotlib is organized in a hierarchy. At the top of the hierarchy is the matplotlib "state-machine environment" which is provided by the matplotlib.pyplot module.

At this level, simple functions are used to add plot elements (lines, images, text, etc.) to the current axes in the current figure.

## Pyplot

matplotlib.pyplot is a collection of command style functions that make matplotlib work like MATLAB. Each pyplot function makes some change to a figure: e.g., 
- creates a figure
- creates a plotting area in a figure
- plots some lines in a plotting area
- decorates the plot with labels, etc.

```python
import matplotlib.pyplot as plt
```

```python
xSeries=[1,2,3,4,3,2]
plt.plot(xSeries)
```

 If you provide a single list or array to the plot() command, matplotlib assumes it is a sequence of y values, and automatically generates the x values for you. Since python ranges start with 0, the default x vector has the same length as y but starts with 0.

```python
xSeries=[1,2,3,4,3,2]
plt.plot(xSeries)
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.title('Sample Graph 1')
```

Matplotlib is not limited to lists, It supports numpy arrays. In fact, all sequences are converted to numpy arrays internally.

```python
xSeries=np.linspace(0,100,30)
ySeries=np.random.rand(30)

plt.plot(xSeries,ySeries)
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.title('Sample Graph 1')
```

## Subplot

```python
plt.figure(figsize=(20, 5))

plt.subplot(131)
plt.bar(kk, kk3)
plt.subplot(132)
plt.scatter(kk, kk3)
plt.subplot(133)
plt.plot(kk, kk3)
plt.suptitle('Subplots with Categorical values')
```

You can create an arbitrary number of subplots and axes.
If you want to place an axes manually, i.e., not on a rectangular grid, use the axes() command, which allows you to specify the location as axes([left, bottom, width, height]) where all values are in fractional (0 to 1) coordinates. See Axes Demo for an example of placing axes manually and Basic Subplot Demo for an example with lots of subplots.

You can create multiple figures by using multiple figure() calls with an increasing figure number. Of course, each figure can contain as many axes and subplots as your heart desires:

```python
plt.figure(figsize=(20,5))
plt.subplot(131)
plt.subplot(132)
plt.subplot(133)
```

```python
plt.figure(figsize=(20,5))
plt.subplot(121)
plt.subplot(122)
# plt.subplot(133)
```

###  (numrows, numcols, plot_number )  >>>>  numrows*numcols <10

```python
plt.figure(figsize=(10,3))
# plt.subplot(11)
plt.subplot(131)
plt.scatter(kk,kk3)
plt.subplot(132)
plt.plot(kk,kk3,'g*')
plt.subplot(133)
plt.plot(kk,kk3,'b*')
```

```python
plt.figure(figsize=(20,10))
# plt.subplot(11)
plt.subplot(231)
plt.scatter(kk,kk3)
plt.subplot(232)
plt.plot(kk,kk3,'g*')
plt.subplot(233)
plt.bar(kk,kk3,'g_')
plt.subplot(234)
plt.plot(kk,kk3,'r|')
plt.subplot(235)
plt.plot(kk,kk3,'bH')
plt.subplot(236)
plt.plot(kk,kk3,'yh')
```

The figure() command here is optional because figure(1) will be created by default, just as a subplot(111) will be created by default if you don't manually specify any axes. The subplot() command specifies numrows, numcols, plot_number where plot_number ranges from 1 to numrows*numcols. The commas in the subplot command are optional if numrows*numcols<10. So subplot(211) is identical to subplot(2, 1, 1).

## Labeling and use of text in plotting

```python
someScores=np.random.random(100)*100
```

```python
meanVal,stdDevVal=someScores.mean(),someScores.std()
```

```python
plt.hist(someScores,facecolor='g', alpha=0.75)
plt.xlabel('Scores')
plt.ylabel('OCCURENCE')
plt.title('Histogram of scores')
plt.text(20, 18, '$\mu={},\ \sigma=${}'.format(meanVal,stdDevVal))
plt.axis([0, 100, 0,22])
plt.grid(True)
```

## Annotate

```python
plt.hist(someScores,facecolor='g', alpha=0.75)
plt.xlabel('Scores')
plt.ylabel('OCCURENCE')
plt.title('Histogram of scores')
plt.text(20, 18, '$\mu={},\ \sigma=${}'.format(meanVal,stdDevVal))
plt.axis([0, 100, 0,22])
plt.grid(True)
plt.annotate('local max', xy=(12, 12), xytext=(20, 12.5),
             arrowprops=dict(facecolor='black', shrink=0.05),
             )
```

```python
xSeries=np.random.random(500)
plt.plot(xSeries)
plt.yscale('logit')
plt.tick_params(direction='out', length=6, width=2, colors='r',
                   grid_color='r', grid_alpha=0.5)
# plt.close()
```

```python
xSeries=np.random.random(500)
# ax,fig=plt.figure()
plt.plot(xSeries)
plt.yscale('logit')
plt.tick_params(
    axis='y',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    left=False,      # ticks along the bottom edge are of
    labelleft=False)
plt.axis()
```

## Save the graph

```python
xSeries=np.random.random(500)
# ax,fig=plt.figure()
plt.plot(xSeries)
plt.yscale('logit')
plt.tick_params(
    axis='y',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    left=False,      # ticks along the bottom edge are of
    labelleft=False)
plt.axis()
plt.savefig('someImage.png')
plt.close()
```

You can clear the current figure with clf() and the current axes with cla() and  close() command clears the memory consumed by the plotting process.
